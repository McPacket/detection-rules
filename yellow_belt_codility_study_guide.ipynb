## 4.1 Creating Lists

## 5.1 String Basics

```python
# Strings are sequences of characters
text = "Hello, World!"

# Indexing (like lists)
print("First char:", text[0])      # H
print("Last char:", text[-1])      # !

# Slicing
print("First word:", text[:5])     # Hello
print("Last word:", text[7:])      # World!
print("Reversed:", text[::-1])     # !dlroW ,olleH

# String properties
print("\nLength:", len(text))
print("'o' in text:", 'o' in text)  # True

# Strings are immutable (can't change)
# text[0] = 'J'  # This would cause an error!
```

## 5.2 String Methods

```python
# Case conversion
text = "Hello World"
print("Upper:", text.upper())       # HELLO WORLD
print("Lower:", text.lower())       # hello world
print("Title:", text.title())       # Hello World

# Whitespace
text = "  Hello  "
print("Stripped:", text.strip())    # "Hello"

# Searching
text = "Python is awesome"
print("\nFind 'is':", text.find('is'))        # 7
print("Count 'o':", text.count('o'))           # 2  
print("Starts with 'Python':", text.startswith('Python'))  # True
print("Ends with 'awesome':", text.endswith('awesome'))    # True

# Replace
print("Replace:", text.replace('awesome', 'great'))

# Split and join
words = text.split()  # Split by whitespace
print("\nWords:", words)

joined = '-'.join(words)
print("Joined:", joined)
```

## 7.1 Searching

```python
# Linear search - find element in list
def find_element(arr, target):
    for i in range(len(arr)):
        if arr[i] == target:
            return i  # Return index
    return -1  # Not found

numbers = [3, 7, 1, 9, 5]
print("Index of 9:", find_element(numbers, 9))
print("Index of 10:", find_element(numbers, 10))

# Find min/max
def find_min(arr):
    min_val = arr[0]
    for num in arr:
        if num < min_val:
            min_val = num
    return min_val

print("Minimum:", find_min(numbers))
```

## 7.2 Sorting

```python
# Using built-in sort
numbers = [3, 7, 1, 9, 5, 2]

# sorted() returns new list
sorted_nums = sorted(numbers)
print("Original:", numbers)
print("Sorted:", sorted_nums)

# .sort() modifies in place
numbers.sort()
print("After .sort():", numbers)

# Sort descending
numbers.sort(reverse=True)
print("Descending:", numbers)
```

## 7.3 Frequency Counting

```python
# Count occurrences
def count_item(arr, target):
    count = 0
    for item in arr:
        if item == target:
            count += 1
    return count

numbers = [1, 2, 3, 2, 4, 2, 5]
print("Count of 2:", count_item(numbers, 2))

# Frequency counter with dictionary
def frequency(arr):
    freq = {}
    for item in arr:
        if item in freq:
            freq[item] += 1
        else:
            freq[item] = 1
    return freq

print("Frequencies:", frequency(numbers))
```

## 7.4 Two-Pointer Technique

```python
# Check palindrome using two pointers
def is_palindrome(s):
    left = 0
    right = len(s) - 1
    
    while left < right:
        if s[left] != s[right]:
            return False
        left += 1
        right -= 1
    
    return True

print("'racecar':", is_palindrome("racecar"))
print("'hello':", is_palindrome("hello"))

# Reverse array in place
def reverse_array(arr):
    left, right = 0, len(arr) - 1
    while left < right:
        arr[left], arr[right] = arr[right], arr[left]
        left += 1
        right -= 1
    return arr

nums = [1, 2, 3, 4, 5]
print("Reversed:", reverse_array(nums))
```
